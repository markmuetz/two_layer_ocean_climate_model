```python
obs_temp_anomaly = pd.read_csv('data/orig/observed_temperature_anomaly_1850-2016.csv')
obs_temp_uncert = pd.read_csv('data/orig/observed_temperature_uncertainty.csv')
```

```python
years = obs_temp_anomaly.YEAR.astype(int).values
obs = obs_temp_anomaly['Obs.'].values
```

```python
years
```

```python
obs
```

```python
obs_temp_uncert.loc[len(obs_temp_uncert)] = (2016, None, None, None)
```
